```python
import numpy as np
sales = np.random.randint(100,301, size=12)
np.save('sales.npy',sales)#guardar el array
```

```python
loades_sales = np.load('sales.npy') #cargar el archivo  generado en la anterior seccion
print(loades_sales)
```

```python
np.savetxt('sales.csv',sales,delimiter=',') #guardar el array en formato csv
```

```python
loaded_sales_csv = np.loadtxt('sales.csv',delimiter=',') #cargar el archivo csv
print(loaded_sales_csv)
```

**MATPLOLIB** es una libreria de python con la que se puede realizar gráficas.

```python
import matplotlib.pyplot as plt

months = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
plt.figure(figsize=(10,5))#se especifica el tamaño de la figura
plt.bar(months, sales, color = 'b', label="Ventas Generadas")
plt.title('Ventas Mensuales')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.xticks(rotation=45)#rotación de las etiquetas en el eje x
plt.show()
```

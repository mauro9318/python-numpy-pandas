Las PIVOT TABLES son una herramienta que permite resumir y reorganizar columnas de datos crudos. Con esta herramienta se puede realizar calculos estad√≠sticos sencillos como sumas, promedios y conteos.

```python
import pandas as pd
sale_data = pd.read_csv("/content/Online_Retail.csv", encoding='latin-1')
sale_data.head()
```

```python
#creaci√≥n de una PIVOT TABLE
pivot_table = pd.pivot_table(sale_data, values='Quantity', index='Country', columns='StockCode', aggfunc='sum')
print(pivot_table)
```

Al trabajar con datos temporales en pandas, es crucial que los datos esten en formatos adecuados. Pandas maneja estas columnas usando el tipo de dato DATETIME.

```python
import pandas as pd

#Cargar el dataset desde el archivo CSV

filePath = '/content/Online_Retail.csv'
df = pd.read_csv(filePath, encoding='latin-1')

#Mostrar las primeras filas del DataFrame

print(df.head())
```

```python
#Obtener información del dataset

df.info()
```

```python
#Como se puede ver, el tipo de dato de la columna InvoiceDate es del tipo objeto
#por tal motivo se debe realizar una conversión

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

#Verificar el tipo de dato de la columna InvoiceDate

print(df['InvoiceDate'].dtype)
```

```python
df.info()
```

```python
#eliminar los valores nulos en la columna InvoiceDate
df.dropna(subset=['InvoiceDate'],inplace=True)
```

```python
#Establecer la columna InvoiceDate como indice
df.set_index('InvoiceDate',inplace=True)
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df['Week'] = df.index.weekday
df['Hour'] = df.index.hour

print(df['Year'])
print(df['Hour'])
```
